### Data Visualization On Differents Concerns

```python
import matplotlib.pyplot as plt
from sklearn import metrics
df.plot(x='Name', y='Views count', style='v')  
plt.title('Views count')  
plt.xlabel('Name')  
plt.xticks(rotation=45)
plt.ylabel('Views count')  
plt.show()
```

### Scatter Plot

```python
df.plot(kind='scatter', x='Name', y='Likes count', figsize=(8, 6), color='darkgreen')
plt.title('Scatter plot for the Name vs Likes count')
plt.xlabel('Name')
plt.xticks(rotation=45)
plt.ylabel('Likes count')
plt.show()
```

```python
#sns.distplot(df['Name'])
import seaborn as sns
sns.distplot(df['Views count'])
```

### Comments Vs Views Count   (using Joint plot)

```python
with sns.axes_style('white'):
    sns.jointplot("Comments", "Views count", df, kind='kde');
```
